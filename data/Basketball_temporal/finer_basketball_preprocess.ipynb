```python
import pandas as pd
import numpy as np
import networkx as nx
import scipy.sparse as sp
from stellargraph import StellarDiGraph
```

```python
data = pd.read_csv('./basketball_temporal_season.csv')
folder_name = './'
```

```python
data
```

```python
data.columns
```

```python
data1 = data[['se2so0', 'wte2m', 'lte2m', 'wscore']].copy()
data1.columns=['season','source', 'target', 'weight']
```

```python
data2 = data[['se2so0', 'lte2m', 'wte2m', 'lscore']].copy()
data2.columns=['season','source', 'target', 'weight']
data = pd.concat([data1,data2], ignore_index=True)
data
```

```python
data['weight'] += 0.1
data
```

```python
for season in range(1985, 2015):
    df = data[data['season']==season][['source', 'target', 'weight']]
    graph = StellarDiGraph(edges=df)
    print(graph.info())
    adj = graph.to_adjacency_matrix(weighted=True)
    assert np.abs(adj-adj.transpose()).sum() != 0
    node_list = list(graph.nodes())
    sp.save_npz('finer'+str(season)+'adj', adj)
    np.save('finer'+str(season)+'node_list', node_list)
    print('Finer'+str(season)+' done!')
```
